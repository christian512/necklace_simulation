```python
from necklace_model import Necklace
from genetic_algorithm import GeneticAlgorithm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Genetica Algorithm on "normal" states
nkl = Necklace(20,2)
ga = GeneticAlgorithm()
ga.set_model(nkl)
energies,energiesVBSF = ga.run(population_size=1000,num_gens=200,clone_rate=0.5,crossover_rate=0.2,mutation_rate=0.5)
```

```python
plt.xlabel('Time')
plt.title('GA on collapsed states')
plt.plot(energies,label='AVG energy')
plt.plot(energiesVBSF,label='BSF energy')
```

```python
# Genetica Algorithm on "expanded" states
nkl = Necklace(20,2)
ga = GeneticAlgorithm()
ga.set_model(nkl)
energies,energiesVBSF = ga.run_expanded(population_size=3000,num_gens=75,clone_rate=0.5,crossover_rate=0.2,mutation_rate=0.5)
```

```python
plt.title('GA on expanded states')
plt.xlabel('Time')
plt.plot(energies,label='AVG energy')
plt.plot(energiesVBSF,label='BSF energy')
```
